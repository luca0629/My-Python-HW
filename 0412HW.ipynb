```python
%matplotlib inline

import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def loss(w):
    return w**2 - w
def d_loss(w):
    return 2*w - 1
```

```python
w = np.linspace(-0.5, 1.5, 1000)
plt.plot(w, loss(w))
plt.title('Loss function=w**2 - w')
plt.xlabel('weight')
plt.ylabel('loss')
```

```python
initial = random.uniform(-0.5, 1.5)
print('initial weight:', initial)
```

```python
w = np.linspace(-0.5, 1.5, 1000)
plt.plot(w, loss(w))
plt.title('Loss function=w**2 - w')
plt.xlabel('weight')
plt.ylabel('loss')

w = initial
lr = 0.1
for i in range(10000):
    plt.scatter(w, loss(w))
    if loss(w) == -0.25:
        print('find minimum at step', i+1)
        break
    else:
        w -= lr * d_loss(w)
```
